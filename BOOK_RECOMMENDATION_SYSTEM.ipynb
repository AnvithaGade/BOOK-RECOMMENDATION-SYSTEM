# Collabrative Filtering

```python
#Merging Books dataset with Ratings
ratings_with_names=ratings.merge(books,on='ISBN')
```

```python
ratings_with_names
```

```python
ratings_with_names.isnull().sum()
```

```python
#Considering users who rated greater than 100 books for better performance
x=ratings_with_names.groupby('User-ID').count()['Book-Rating']>100
bookworms=x[x].index
```

```python
x
```

```python
refined_ratings=ratings_with_names[ratings_with_names['User-ID'].isin(bookworms)]
```

```python
#cosidering famous books with ratings greater than 50(given by users)
y=refined_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
y
```

```python
#overall ratings
final_ratings=refined_ratings[refined_ratings['Book-Title'].isin(famous_books)]
```

```python
#Creating a table with ratings for books given by various users
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.head()
```

```python
pt.describe()
```

```python
pt.info()
```

```python
#filling null values(the books that have no ratings by various users)
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
#importing cosine_similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
#Defining a method to retrieve similar book recommendatations
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        data.append(item)
    return data
    
```

```python
#Taking user input for Book-Title
book_title=input("Enter a Book-Title: ")
```

```python
#printing Top 5 Book Recommendations for the specified Book
print('\033[1m' +"BOOK_TITLE , BOOK_AUTHOR")
recommend(book_title)
```
